```python
df= pd.read_csv('saturated_gene_names.csv')
b= df.iloc[:, 1]
df1= pd.read_csv('saturated_gene_names anaerobic.csv')
c= df1.iloc[:, 1]
df2= pd.read_csv('saturated_gene_names study conditions.csv')
d= df2.iloc[:, 1]
df3= pd.read_csv('saturated_gene_names M9 media.csv')
e= df3.iloc[:, 1]
df4= pd.read_csv('saturated_gene_names lactose.csv')
f= df4.iloc[:, 1]

union = pd.Series(np.union1d(b, c))
intersect = pd.Series(np.intersect1d(b, c))
bigb2= union[~union.isin(intersect)]

union = pd.Series(np.union1d(b, d))
intersect = pd.Series(np.intersect1d(b, d))
bigb3= union[~union.isin(intersect)]

union = pd.Series(np.union1d(b, e))
intersect = pd.Series(np.intersect1d(b, e))
bigb4= union[~union.isin(intersect)]

union = pd.Series(np.union1d(b, f))
intersect = pd.Series(np.intersect1d(b, f))
bigb5= union[~union.isin(intersect)]

D= pd.DataFrame(bigb2, columns=['Reaction names'])
D.to_csv("Differences saturated anaerobic")
D1= pd.DataFrame(bigb3, columns=['Reaction names'])
D1.to_csv("Differences saturated study conditions")
D2= pd.DataFrame(bigb4, columns=['Reaction names'])
D2.to_csv("Differences saturated M9 media")
D3= pd.DataFrame(bigb5, columns=['Reaction names'])
D3.to_csv("Differences saturated lactose")
```
